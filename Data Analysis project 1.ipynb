```python
import pandas as pd
```

```python
inventory  = pd.read_csv("PandasAssignments/Assignment2/inventory.csv")
inventory
```

```python
staten_islands = inventory.head(10)
staten_islands
```

```python
product_request = staten_islands.product_description
product_request
```

```python
seed_requests = inventory[(inventory.location=='Brooklyn')&(inventory.product_type=='seeds')]['product_description'].values
seed_requests
```

```python
inventory['in_stock'] = inventory.quantity.apply(lambda x: True if x>0 else False )
inventory.sample(5)
```

```python
inventory['total_value'] = inventory.price * inventory.quantity
inventory
```

```python
inventory['complete_description'] = inventory.product_type+' - '+inventory.product_description
inventory.head()
```

```python
inventory['full_description'] = inventory.apply(lambda row: f"{row.product_type}-{row.product_description}",axis=1)
```
